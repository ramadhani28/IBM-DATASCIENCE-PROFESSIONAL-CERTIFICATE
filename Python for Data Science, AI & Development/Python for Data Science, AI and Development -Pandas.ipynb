### loc is used for label-based indexing, meaning you select rows and columns by their actual names or labels, including custom index values and column headers. In contrast, iloc is used for integer-location based indexing, allowing you to select rows and columns strictly by their numerical positions (starting from 0), regardless of their labels. A key difference in slicing is that loc includes the end label in a slice, while iloc excludes the end position, mimicking standard 



```python
# Using loc (actual names or labels, including custom index values and column headers)
df.loc[2,'Salary']
```

```python
# Using iloc (integer-location-based indexing)
df.iloc[2,3]
```
